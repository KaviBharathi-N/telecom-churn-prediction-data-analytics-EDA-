```python
d=pd.read_csv('Telecom Churn (1).csv')
```

```python
import matplotlib.pyplot as plt


data = d['Churn'].value_counts()

explode = (0, 0.2)
colors = ["blue", "red"]

plt.pie(data, explode=explode, autopct='%1.1f%%', shadow=True, radius=2.0, labels=["Not Churned Customer", "Churned Customer"], colors=colors)

circle = plt.Circle((0, 0), 1, color='white')

p = plt.gcf()
p.gca().add_artist(circle)

plt.title("Donut Plot")

plt.show()

```

```python
sns.set(style='darkgrid')
ax=sns.countplot(x='Area code',hue="Churn",data=d)
plt.show()
```

```python
churn_df=d[d["Churn"]==bool(True)]
not_churn_df=d[d["Churn"]==bool(False)]
```

```python
sns.distplot(d["Account length"])
```

```python
sns.distplot(d["Account length"],color='yellow',label='All')
sns.distplot(churn_df["Account length"],color='red', hist=False,label='Churned')
sns.distplot(not_churn_df["Account length"],color='green', hist=False,label='not Churned')
plt.legend()
```
